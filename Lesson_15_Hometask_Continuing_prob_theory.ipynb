«Параллелограмм», равна 0,6. В сборнике нет задач, которые
одновременно относятся к этим двум темам. Найдите вероятность
того, что на экзамене школьнику достанется задача по одной из
этих двух тем.

```python
import numpy as np
import scipy.stats as sps
from scipy import stats
from scipy.special import perm, comb, factorial
```

```python
# Задача 1.

answer = 5 / 1000 * 100
print(f'1. Вероятность выбрать бракованный телевизор: {answer} %')

# Задача 2.

answer = 6 / (9+6+5) * 100
print(f'2. Вероятность достать желтый шар: {answer} %')

# Задача 3.

'''
Решением будет указать формулу расчёта вероятности, поскольку мы не знаем, какой по счёту будет тянуть номерок сборная Португалии.
Расчитаем частный вариант, если сборная Португалии тянет номерок 6-ой, когда уже 2 члена II-ой группы определились:
'''
N = 4
K = 16
n = 2
k = 5
answer = round((N-n) / (K-k) * 100, 2)
print(f'3. Вероятность равна: P = (N-n) / (K-k). В частном случае, при n = {n}, k = {k}: {answer} %')

# Задача 4.
'''
Аналогично предыдущей, в самом начале шанс 6/8, с вводными расчёт по формуле.
'''
N = 6
K = 8
n = 3
k = 3
answer = round((N-n) / (K-k) * 100, 2)
print(f'4. Вероятность равна: P = (N-n) / (K-k). В частном случае, при n = {n}, k = {k}: {answer} %')

# Задача 5.
'''
Задача обратная задачам 3-4, значит неблагоприятная P = 1 - благоприятная P.
'''
N = 4
K = 16
n = 3
k = 8
answer = 100 - round((N-n) / (K-k) * 100, 2)
print(f'5. Неблаг.вероятность попасть в группу А: P = 1 - (N-n) / (K-k). В частном случае, при n = {n}, k = {k}: {answer} %')

# Задача 6.
'''
Предположим, Аня уже попала в какую-то группу, соответственно в этой группе осталось 2 свободных места.
Тогда вероятность Нине попасть в группу к Ане равна (3-1) / (21-1) = 10%
'''
answer = (3-1) / (21-1) * 100
print(f'6. Благ.вероятность попасть в одну группу: {answer} %')

# Задача 7.
'''
В часах 12 интервалов. Интервалов между VII и I делением: (12+1) - 7 = 6 интервалов.
'''
answer = (13 - 7) / 12 * 100
print(f'7. Вероятность остановки стрелки между VII и I: {answer} %')

# Задача 8.
'''
Аналогично предыдущей задаче. Поскольку в условии говорится "...достигнув отметки 6...", соответственно говорится о ходе по часовой стрелке.
Интервалов между VI и IX делением: 9 - 6 = 3 интервала.
'''
answer = (9 - 6) / 12 * 100
print(f'8. Вероятность остановки стрелки между VI и XI: {answer} %')

# Задача 9.
'''
Всего вариантов выпадения: 2**2:
1) О О
2) О Р
3) Р О
4) Р Р
Из здравой логики, пример достаочно простой, чтобы понять во скольких случаях будет только 1 орёл (или решка, 50/50).
Однако в общем случае, нам было бы необходимо воспользоваться формулой биноминального распределения.
'''
n = 2
k = 1
p = 0.5
answer = round(sps.binom.pmf(k, n, p) * 100, 2)
print(f'9. Вероятность выпадения ровно {k} орлов из {n} попыток: {answer} %')

# Задача 10.
'''
В этой задаче также воспользуемся биноминальным распределением.
'''
n = 2
k = 0
p = 0.5
answer = round(sps.binom.pmf(k, n, p) * 100, 2)
print(f'10. Вероятность выпадения ровно {k} орлов из {n} попыток: {answer} %')

# Задача 11.

answer = (1-0.19) * 100
print(f'11. Вероятность, что ручка пишет хорошо: {answer} %')

# Задача 12.
'''
Союз "или" говорит нем о дизъюнкции - сложении вероятностей. Соответственно, ответом будет общая вероятность событий.
'''
answer = round((1-0.965) * 100, 2)
print(f'12. Вероятность выхода за пределы 0.01 мм отклонения диаметра подшипника: {answer} %')

# Задача 13.
'''
Также видим дизъюнкцию. Сложим вероятности.
'''
answer = round((0.6+0.1) * 100, 2)
print(f'13. Вероятность попадения любой из 2 задач: {answer} %')
```
